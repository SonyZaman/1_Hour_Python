```python
def factorial(number):
  if number == 0:
    return 1
  else:
    result = 1
    for i in range(1,number+1):
      result*=i
    return result


print(factorial(5))
```

```python
def isPrime(number):
  if number == 1:
    return False
  if number % 2 ==0:
    return False

  for i in range(3,number,2):
    if number%i ==0:
      return False

  return True

print(isPrime(11))
```

## Set

```python
my_set = {1, 2, 3, 4, 5}
print(my_set)
print()

# Adding elements
my_set.add(6)
my_set.add(2) # Adding an existing element has no effect
print(my_set)
print()

# Removing elements
my_set.remove(3) # Removes a specified element
print(my_set)
print()

# Discarding elements (safer, doesn't raise error if element not found)
my_set.discard(1)
my_set.discard(99) # No error if 99 is not in set
print(my_set)
print()

# Set operations
other_set = {4, 5, 6, 7, 8}

# Union
union_set = my_set.union(other_set)
print(union_set)
print()

# Intersection
intersection_set = my_set.intersection(other_set)
print(intersection_set)
print()

# Difference
difference_set = my_set.difference(other_set)
print(difference_set)
print()

# Symmetric Difference
symmetric_difference_set = my_set.symmetric_difference(other_set)
print(symmetric_difference_set)
print()

# Check if an element is in the set
print(4 in my_set)
print(10 in my_set)
```

# Object Oriented Programming

Static Methods and Class Variables


```python
class MathUtils:
    # Class variable (static property): belongs to the class itself, not instances
    PI = 3.14159

    # Static method: does not depend on an instance of the class (no 'self' parameter)
    # It can be called directly on the class itself.
    @staticmethod
    def add(x, y):
        return x + y

    @staticmethod
    def circle_area(radius):
        # Static methods can access class variables using the class name
        return MathUtils.PI * radius * radius

# Accessing class variable directly from the class
print(MathUtils.PI)

# Calling static methods directly from the class
print(MathUtils.add(5, 3))
print(MathUtils.circle_area(2))

# We can also call static methods through an instance, but it's less common
math_obj = MathUtils()
print(math_obj.add(10, 20))
```

# Classes and Objects

```python
class Dog:
    # Class attribute: shared by all instances of the class
    species = "Canis familiaris"

    # Constructor method (__init__): initializes new objects
    def __init__(self, name, age):
        self.name = name  # Instance attribute: unique to each instance
        self.age = age    # Instance attribute

    # Instance method: operates on the instance's attributes
    def bark(self):
        return f"{self.name} says Woof!"

    def get_age(self):
        return self.age

# Create objects (instances) of the Dog class
my_dog = Dog("Buddy", 3)
your_dog = Dog("Lucy", 5)

# Access instance attributes
print(my_dog.name)
print(your_dog.age)

# Call instance methods
print(my_dog.bark())
print(your_dog.get_age())

# Access class attribute
print(Dog.species)
print(my_dog.species)
```

# Inheritance


```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        return f"Hi, my name is {self.name} and I am {self.age} years old."

class Student(Person):
    def __init__(self, name, age, student_id, major):

        super().__init__(name, age)
        self.student_id = student_id
        self.major = major

    def study(self):
        return f"{self.name} (ID: {self.student_id}) is studying {self.major}."

    def introduce(self): # Overriding the parent method for specific behavior
        return f"Hello, I'm {self.name}, a student majoring in {self.major} (ID: {self.student_id})."

class Teacher(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject

    def teach(self):
        return f"{self.name} is teaching {self.subject}."

# Create objects of parent and child classes
person1 = Person("Alice", 30)
student1 = Student("Bob", 20, "S123", "Computer Science")
teacher1 = Teacher("Mr. Smith", 45, "Mathematics")

print(person1.introduce())
print(student1.introduce()) # Calls the overridden method in Student
print(student1.study())
print(teacher1.introduce())
print(teacher1.teach())
```

# Polymorphism

**Polymorphism** means "many forms." In OOP, it allows objects of different classes to be treated as objects of a common superclass. This means that a single interface can be used to represent different underlying forms (data types or classes), and a method can behave differently depending on the object it's called on.

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        # Raising NotImplementedError ensures subclasses provide their own implementation.
        raise NotImplementedError("Subclass must implement abstract method 'speak'")

class Dog(Animal):
    def speak(self):
        return f"{self.name} says Woof!"

class Cat(Animal):
    def speak(self):
        return f"{self.name} says Meow!"

class Duck(Animal):
    def speak(self):
        return f"{self.name} says Quack!"

# A function demonstrating polymorphism: it can take any Animal object
# and call its 'speak' method, regardless of the specific subclass type.
def make_animal_speak(animal):
    print(animal.speak())

# Create different animal objects
dog = Dog("Buddy")
cat = Cat("Whiskers")
duck = Duck("Daffy")

# Demonstrate polymorphism by passing different types of Animal objects
# to the same function.
print("--- Polymorphism with a function ---")
make_animal_speak(dog)
make_animal_speak(cat)
make_animal_speak(duck)

# Another way to show polymorphism with a list of heterogeneous objects
print("\n--- Polymorphism with a list ---")
animals = [Dog("Max"), Cat("Luna"), Duck("Donald")]
for animal in animals:
    # The correct 'speak' method is called based on the object's actual type at runtime.
    print(animal.speak())
```

# Encapsulation

**Encapsulation** is the bundling of data (attributes) and methods (functions) that operate on the data into a single unit, or class. It also involves restricting direct access to some of an object's components, typically by making them 'private' or 'protected', and providing public methods to interact with them, thus protecting data integrity.

```python
class BankAccount:
    def __init__(self, account_holder, initial_balance):
        self.account_holder = account_holder  # Public attribute
        # Protected member (by convention): indicated by a single underscore
        # Should be treated as internal to the class or its subclasses.
        self._account_number = "123456789"
        # Private member (name mangling): indicated by double underscores
        # Python renames this attribute to _ClassName__attribute, making it harder to access directly.
        self.__balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}. New balance: {self.__balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew {amount}. New balance: {self.__balance}")
        else:
            print("Invalid withdrawal amount or insufficient funds.")

    # Public method to safely access the private balance
    def get_balance(self):
        return self.__balance



account = BankAccount("Jane Doe", 1000)

# Accessing public attribute
print(account.account_holder)

# Attempting to access protected member (possible but discouraged by convention)
print(account._account_number)

# Attempting to access private member directly will raise an AttributeError
# print(account.__balance)

# Accessing private member via name mangling (not recommended for direct access)
print(account._BankAccount__balance)

# Use public methods to interact with the balance (encapsulation in action)
account.deposit(200)
account.withdraw(150)
account.withdraw(1500) # This will fail due to insufficient funds

print(account.get_balance())
```

# Abstraction



```python
from abc import ABC, abstractmethod

class Shape(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

    # Concrete method: a method with an implementation in the base class
    def describe(self):
        return f"This is a {self.name} shape."

# Concrete subclass: must implement all abstract methods from its ABC
class Circle(Shape):
    def __init__(self, radius):
        super().__init__("Circle")
        self.radius = radius

    def area(self):
        return 3.14159 * self.radius * self.radius

    def perimeter(self):
        return 2 * 3.14159 * self.radius

# Concrete subclass
class Rectangle(Shape):
    def __init__(self, width, height):
        super().__init__("Rectangle")
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

# We cannot instantiate an abstract class directly
# generic_shape = Shape("Generic") # This would raise a TypeError

circle = Circle(5)
rectangle = Rectangle(4, 6)

print(circle.describe())
print(f"{circle.area():.2f}")
print(f"{circle.perimeter():.2f}")

print(rectangle.describe())
print(f"{rectangle.area():.2f}")
print(f"{rectangle.perimeter():.2f}")
```
